```python
import numpy as np
import os # Import os to check for file existence

def high_dimensional_optimization(objective_function, initial_inputs_file, initial_outputs_file, perturbation_scale=0.05, max_iterations=100):
    """Performs high-dimensional optimization using a local search approach."""
    if not os.path.exists(initial_inputs_file):
        raise FileNotFoundError(f"Error: File not found - {initial_inputs_file}")
    if not os.path.exists(initial_outputs_file):
        raise FileNotFoundError(f"Error: File not found - {initial_outputs_file}")

    initial_inputs = np.load(initial_inputs_file)
    initial_outputs = np.load(initial_outputs_file)

    best_initial_index = np.argmin(initial_outputs)
    initial_point = initial_inputs[best_initial_index]

    current_point = initial_point
    best_point = current_point
    best_value = objective_function(current_point)

    for _ in range(max_iterations):
        neighbor_point = current_point + np.random.uniform(-perturbation_scale, perturbation_scale, size=current_point.shape)
        neighbor_point = np.clip(neighbor_point, 0, 1)
        neighbor_value = objective_function(neighbor_point)

        if neighbor_value < best_value:
            best_value = neighbor_value
            best_point = neighbor_point
            current_point = neighbor_point

    return best_point, best_value

def my_objective_function(inputs):
    """Example objective function (sum of squares)."""
    # Replace this with your actual 8-dimensional objective function
    return np.sum(inputs**2)

def generate_next_query(current_input, perturbation_scale=0.05, num_queries=5):
    """Generates multiple next queries by adding small random perturbations."""
    next_queries = []
    for _ in range(num_queries):
        perturbations = np.random.uniform(-perturbation_scale, perturbation_scale, size=current_input.shape)
        next_query = current_input + perturbations
        next_query = np.clip(next_query, 0, 1)
        next_queries.append(next_query)
    return next_queries

# Run optimization using the updated data
best_point, best_value = high_dimensional_optimization(
    my_objective_function,
    initial_inputs_file='/content/updated_inputs.npy',
    initial_outputs_file='/content/updated_outputs.npy'
)

print("Best point found by optimization:", best_point)
print("Best value found by optimization:", best_value)

# Generate next queries around the best point found by optimization
next_queries = generate_next_query(best_point)
print("\nNext Queries generated from the best point:")
for query in next_queries:
    print(query)
```
